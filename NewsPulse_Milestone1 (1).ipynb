```python
import requests
import pandas as pd
```

```python
api_key = "6e1249eec94d4aec96f9e622346bc9fb"
```

```python
url = f"https://newsapi.org/v2/everything?q=india&pageSize=20&apiKey={api_key}"
```

```python
response = requests.get(url)
data = response.json()
```

```python
print("status:",data['status'])
print("total result:",data.get("total results",0))
```

```python
articles = data.get("articles",[])
for i,article in enumerate(articles,start=1):
  print(f"article{i}")
  print("title:",article.get("title"))
  print("source:",article.get("source",{}).get("name"))
  print("published date:",article.get("publishedAt"))
  print("-"* 40)
```

```python
news_list = []
for article in articles:
    news_list.append({
        "Title": article.get("title"),
        "Description": article.get("description"),
        "Source": article.get("source", {}).get("name"),
        "Date": article.get("publishedAt")
    })
    df=pd.DataFrame(news_list)
    df.to_csv("news_data.csv",index=False)
    print("news_data.csv created successfully!")
```

```python
df.head()
```

```python
import re

# Remove empty rows
df_clean = df.dropna()

# Remove duplicate titles
df_clean = df_clean.drop_duplicates(subset="Title")

# Clean special characters
def clean_text(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', str(text)).strip()

df_clean["Title"] = df_clean["Title"].apply(clean_text)
df_clean["Description"] = df_clean["Description"].apply(clean_text)

df_clean.to_csv("news_data_cleaned.csv", index=False)

print("news_data_cleaned.csv created successfully!")
```

```python
print("Total News Articles:", len(df_clean))
print("Unique News Sources:", df_clean["Source"].nunique())
```
