```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("news_with_sentiment (1).csv")
df.head()
```

```python
df = df.drop_duplicates(subset=['Title'])
```

```python
df = df.dropna()
```

```python
all_words = " ".join(df["Description"]).lower().split()

word_count = Counter(all_words)

top_words = word_count.most_common(10)

words = [word[0] for word in top_words]
counts = [word[1] for word in top_words]

print(top_words)
```

```python
plt.figure()

plt.bar(words, counts)

plt.xticks(rotation=45)

plt.title("Top 10 Trending Keywords")

plt.xlabel("Keywords")

plt.ylabel("Frequency")

plt.show()
```

```python
sentiment_count = df["Sentiment"].value_counts()

labels = sentiment_count.index

values = sentiment_count.values

print(sentiment_count)
```

```python
plt.figure()

plt.bar(labels, values)

plt.title("Sentiment Distribution")

plt.xlabel("Sentiment")

plt.ylabel("Number of News")

plt.show()
```

```python
plt.figure()

plt.pie(values, labels=labels, autopct='%1.1f%%')

plt.title("Sentiment Distribution Percentage")

plt.show()
```

```python
print("Total News Articles:", len(df))

print("Top Trending Keyword:", words[0])

print("Most Common Sentiment:", sentiment_count.idxmax())
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

st.title("NewsPulse Dashboard")
st.write("Global News Trend Analyzer")

df = pd.read_csv("news_with_sentiment (1).csv")

st.write("Dataset Preview")
st.dataframe(df.head())

# Show columns
st.write("Columns in dataset:", df.columns)

# Select a text column
text_column = df.columns[0]

all_words = " ".join(df[text_column].astype(str)).split()
word_count = Counter(all_words)

top_words = word_count.most_common(10)

words = [w[0] for w in top_words]
counts = [w[1] for w in top_words]

st.subheader("Top Trending Keywords")

fig, ax = plt.subplots()
ax.bar(words, counts)
plt.xticks(rotation=45)

st.pyplot(fig)

# Sentiment chart
if "sentiment" in df.columns:
    sentiment_count = df["sentiment"].value_counts()

    labels = sentiment_count.index
    values = sentiment_count.values

    st.subheader("Sentiment Distribution")

    fig2, ax2 = plt.subplots()
    ax2.pie(values, labels=labels, autopct="%1.1f%%")

    st.pyplot(fig2)
```
